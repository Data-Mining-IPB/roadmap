# 1. 🔀 If-Else (Pengambilan Keputusan)

**If-Else** adalah struktur kontrol yang memungkinkan program untuk membuat keputusan berdasarkan kondisi tertentu. Ini seperti cara manusia berpikir: "Jika A benar, maka lakukan X, kalau tidak lakukan Y".

## 🧠 Konsep Dasar:
- **`if`**: Jika kondisi benar (True), jalankan blok kode
- **`elif`** (else if): Jika kondisi pertama salah, cek kondisi kedua  
- **`else`**: Jika semua kondisi salah, jalankan blok ini

## 📝 Sintaks Dasar:
```python
if kondisi1:
    # Kode yang dijalankan jika kondisi1 True
elif kondisi2:
    # Kode yang dijalankan jika kondisi2 True
else:
    # Kode yang dijalankan jika semua kondisi False
```

**Catatan Penting:**
- Gunakan `:` (titik dua) setelah setiap kondisi
- **Indentasi** (spasi) sangat penting di Python - menunjukkan blok kode
- Operator perbandingan: `==` (sama dengan), `!=` (tidak sama), `>`, `<`, `>=`, `<=`

Contohnya seperti memilih jalur berdasarkan nilai siswa:

````python
nilai = 75
if nilai >= 85:
    print("Lulus dengan predikat A")
elif nilai >= 70:
    print("Lulus dengan predikat B")
else:
    print("Belum lulus")
````

### 💡 Penjelasan Kode:
1. **`nilai = 75`** → Mendeklarasikan variabel nilai dengan angka 75
2. **`if nilai >= 85:`** → Cek apakah nilai lebih besar atau sama dengan 85
3. **`elif nilai >= 70:`** → Jika kondisi pertama salah, cek apakah nilai >= 70
4. **`else:`** → Jika kedua kondisi di atas salah, jalankan blok ini

**Urutan pengecekan:** Python memeriksa kondisi dari atas ke bawah. Begitu menemukan kondisi yang benar, blok kode tersebut dijalankan dan kondisi selanjutnya diabaikan.

### 🌡️ Contoh Penerapan Lain: Temperature Converter

Kita juga bisa menggunakan if-else untuk membuat konverter suhu:

````python
suhu_celsius = 12
if suhu_celsius <= 0:
    print(f"{suhu_celsius}°C - Beku! ❄️")
elif suhu_celsius <= 15:
    print(f"{suhu_celsius}°C - Dingin 🥶")
elif suhu_celsius <= 25:
    print(f"{suhu_celsius}°C - Sejuk 😊")
elif suhu_celsius <= 35:
    print(f"{suhu_celsius}°C - Hangat ☀️")
else:
    print(f"{suhu_celsius}°C - Panas! 🔥")
````

### 🔍 Analisis Kode Temperature Converter:
- **f-string** (`f"{variabel}"`) untuk menggabungkan teks dan variabel
- **Emoji** membuat output lebih menarik dan mudah dipahami
- **Logika bertingkat** dari suhu terdingin ke terpanas
- **Operator `<=`** untuk "kurang dari atau sama dengan"

**Tips:** Urutan kondisi penting! Kita mulai dari suhu terendah karena jika dibalik, semua suhu akan masuk kategori pertama.

# 2. 🎯 Match Case (Switch Versi Python)

**Match Case** adalah fitur baru di Python 3.10+ yang mirip dengan `switch` di bahasa pemrograman lain. Ini berguna ketika kita perlu memeriksa satu variabel terhadap banyak nilai yang berbeda.

## 🔄 Perbedaan Match vs If-Else:
- **If-Else**: Baik untuk kondisi kompleks dengan operator perbandingan
- **Match**: Lebih clean dan readable untuk pengecekan nilai eksak

## 📝 Sintaks Dasar:
```python
match variabel:
    case nilai1:
        # Kode jika variabel == nilai1
    case nilai2 | nilai3:  # Bisa gabung beberapa nilai dengan |
        # Kode jika variabel == nilai2 ATAU nilai3
    case _:  # Default case (seperti else)
        # Kode jika tidak ada yang cocok
```

**Keunggulan Match:**
- Lebih mudah dibaca untuk banyak pilihan
- Bisa menggabungkan beberapa nilai dalam satu case dengan `|`
- Pattern matching yang powerful

Contohnya:

````python
hari = "Senin"
match hari:
    case "Senin":
        print("Hari kerja pertama")
    case "Sabtu" | "Minggu":
        print("Akhir pekan!")
    case _:
        print("Hari biasa")
````

### 💡 Penjelasan Kode Match:
1. **`match hari:`** → Mulai pengecekan terhadap variabel `hari`
2. **`case "Senin":`** → Jika hari sama dengan "Senin"
3. **`case "Sabtu" | "Minggu":`** → Jika hari = "Sabtu" ATAU "Minggu" (operator `|`)
4. **`case _:`** → Default case (underscore = wildcard), seperti `else` di if-statement

**Catatan:** `case _:` akan menangkap semua nilai yang tidak cocok dengan case sebelumnya.

# 3. 🔁 While Loops (Perulangan Berdasarkan Kondisi)

**While Loop** adalah perulangan yang berjalan selama kondisi tertentu masih benar (True). Ini seperti perintah: "Selama X masih benar, terus lakukan Y".

## ⚠️ Hal Penting dalam While Loop:
1. **Kondisi** harus eventually menjadi False, kalau tidak loop akan berjalan selamanya (infinite loop)
2. **Update variabel** di dalam loop agar kondisi bisa berubah
3. **Hati-hati dengan infinite loop** - bisa membuat program hang!

## 📝 Sintaks Dasar:
```python
while kondisi:
    # Kode yang diulang
    # PENTING: Update variabel yang mempengaruhi kondisi
```

## 🎯 Kapan Gunakan While:
- Ketika tidak tahu berapa kali loop akan berjalan
- Ketika loop bergantung pada kondisi dinamis
- Untuk validasi input user
- Untuk game loops atau simulasi

Contoh: Menghitung mundur dari 5 ke 1

````python
count = 5
while count > 0:
    print(count)
    count -= 1
print("Selesai!")
````

### 💡 Penjelasan Kode Countdown:
1. **`count = 5`** → Inisialisasi variabel counter
2. **`while count > 0:`** → Selama count lebih besar dari 0
3. **`print(count)`** → Cetak nilai count saat ini
4. **`count -= 1`** → Kurangi count sebesar 1 (equivalent dengan `count = count - 1`)
5. **Loop berakhir** ketika count menjadi 0 (kondisi `count > 0` menjadi False)

**Operator Assignment:**
- `+=` : tambah dan assign (`x += 1` sama dengan `x = x + 1`)
- `-=` : kurang dan assign
- `*=` : kali dan assign  
- `/=` : bagi dan assign

### 🎯 Contoh Penerapan Lain: Simple Guessing Game

Contoh while loop untuk permainan tebak angka sederhana:

````python
target = 7
tebakan = 0
percobaan = 0

while tebakan != target:
    percobaan += 1
    tebakan = int(input(f"Percobaan {percobaan}: Tebak angka 1-10: "))

    if tebakan < target:
        print("Terlalu kecil! 📉")
    elif tebakan > target:
        print("Terlalu besar! 📈")
    else:
        print(f"🎉 Benar! Angka {target} dalam {percobaan} percobaan!")
````

### 🔍 Analisis Kode Guessing Game:
**Bagian 1 - Interactive Version:**
- **`while tebakan != target:`** → Loop hingga tebakan sama dengan target
- **`int(input())`** → Konversi input string ke integer
- **`percobaan += 1`** → Tracking jumlah percobaan
- **Nested if-elif-else** di dalam while untuk memberikan feedback

**Konsep Penting:** Loop akan terus berjalan sampai user menebak dengan benar atau kondisi break terpenuhi.

# 4. 🔁 For Loops (Perulangan Berdasarkan Data)

**For Loop** digunakan untuk iterasi (mengulang) melalui sequence data seperti list, string, range, dll. Berbeda dengan while yang bergantung kondisi, for loop bergantung pada data yang di-iterasi.

## 🎯 Kapan Gunakan For Loop:
- Ketika tahu persis berapa kali loop akan berjalan
- Ketika ingin memproses setiap item dalam collection (list, string, etc.)
- Untuk pattern atau range angka tertentu

## 📝 Sintaks Dasar:
```python
for variable in iterable:
    # Kode yang diulang untuk setiap item
```

## 🗂️ Jenis Data yang Bisa Di-iterasi:
- **String** → setiap karakter
- **List** → setiap elemen  
- **Range** → sequence angka
- **Dictionary** → setiap key
- **Tuple** → setiap elemen

## 🔧 Fungsi Berguna:
- **`range(start, stop, step)`** → Generate sequence angka
- **`enumerate(list)`** → Dapat index dan value sekaligus
- **`len()`** → Panjang data

Contoh: Menampilkan semua huruf dalam string

````python
teks = "Python"
for huruf in teks:
    print(huruf)
````

### 💡 Penjelasan Kode String Iteration:
1. **`teks = "Python"`** → String dengan 6 karakter
2. **`for huruf in teks:`** → Untuk setiap karakter dalam string teks
3. **`huruf`** → Variable temporary yang berisi karakter saat ini
4. **Loop berjalan 6 kali** → Satu kali untuk setiap karakter (P-y-t-h-o-n)

**Konsep:** String dalam Python adalah sequence karakter, jadi bisa di-iterasi seperti list.

````python
belanjaan = ["beras", "gula", "telur", "minyak", "kopi"]
for idx, item in enumerate(belanjaan):
    print(f"{idx}. {item}")
````

### 💡 Penjelasan Kode List Iteration:
1. **`belanjaan = ["beras", "gula", "telur", "minyak", "kopi"]`** → Membuat list berisi nama-nama barang belanjaan.
2. **`for idx, item in enumerate(belanjaan, 1):`** → Melakukan iterasi pada setiap item di list, sekaligus mendapatkan nomor urut mulai dari 1.
3. **`print(f"{idx}. {item}")`** → Mencetak nomor urut dan nama barang pada setiap baris.

**Konsep:**  
- Fungsi `enumerate()` memudahkan mendapatkan index dan value sekaligus saat melakukan loop pada list.
- Cocok untuk membuat daftar bernomor secara otomatis dari data list.

### 📊 Contoh Penerapan Lain: Analisis Data Sederhana

For loop sangat berguna untuk menganalisis data dalam list:

````python
# Analisis nilai siswa
nilai_siswa = [85, 92, 78, 96, 67, 88, 91, 73, 82, 95]

# Hitung rata-rata
total = 0
for nilai in nilai_siswa:
    total += nilai
rata_rata = total / len(nilai_siswa)
print(f"Rata-rata: {rata_rata:.1f}")

# Cari nilai tertinggi dan terendah
tertinggi = nilai_siswa[0]
terendah = nilai_siswa[0]

for nilai in nilai_siswa:
    if nilai > tertinggi:
        tertinggi = nilai
    if nilai < terendah:
        terendah = nilai

print(f"Nilai tertinggi: {tertinggi}")
print(f"Nilai terendah: {terendah}")

# Hitung jumlah siswa yang lulus (>=75)
lulus = 0
for nilai in nilai_siswa:
    if nilai >= 75:
        lulus += 1
print(f"Siswa yang lulus: {lulus} dari {len(nilai_siswa)}")
````

### 🔍 Analisis Kode Data Analysis:

**1. Menghitung Rata-rata:**
- **`total = 0`** → Inisialisasi accumulator
- **`for nilai in nilai_siswa:`** → Iterasi setiap nilai dalam list
- **`total += nilai`** → Tambahkan setiap nilai ke total
- **`rata_rata = total / len(nilai_siswa)`** → Bagi total dengan jumlah data
- **`:.1f`** → Format decimal 1 angka di belakang koma

**2. Mencari Min/Max:**
- **Inisialisasi** dengan elemen pertama list
- **Bandingkan** setiap elemen dengan nilai saat ini
- **Update** jika ditemukan nilai yang lebih tinggi/rendah

**3. Counting dengan Kondisi:**
- **Counter variable** (`lulus = 0`) untuk hitung jumlah
- **Conditional increment** hanya jika memenuhi syarat (>=75)

**Tips:** For loop sangat powerful untuk data processing karena bisa kombinasi dengan if-else!

## 🚀 Konsep Lanjutan

### 🔄 Nested Loops (Loop dalam Loop)
Kita bisa membuat loop di dalam loop untuk pola yang lebih kompleks:

````python
# Membuat tabel perkalian 5x5
print("Tabel Perkalian:")
for i in range(1, 6):
    for j in range(1, 6):
        hasil = i * j
        print(f"{hasil:2}", end=" ")
    print()  # pindah baris

print("\n" + "="*20)

# Membuat pola berlian dengan bintang
print("Pola Berlian:")
for i in range(1, 4):
    print(" " * (3-i) + "*" * (2*i-1))
for i in range(2, 0, -1):
    print(" " * (3-i) + "*" * (2*i-1))
````

### 🔍 Analisis Kode Nested Loops:

**1. Tabel Perkalian:**
- **Outer loop** (`for i in range(1, 6)`) → Baris (1-5)
- **Inner loop** (`for j in range(1, 6)`) → Kolom (1-5)  
- **`print(f"{hasil:2}", end=" ")`** → Format 2 digit, tanpa newline
- **`print()`** → Newline setelah selesai satu baris
- **Total iterasi:** 5 × 5 = 25 kali

**2. Pola Berlian:**
- **Bagian atas:** Loop naik (1-3) untuk membuat segitiga
- **Bagian bawah:** Loop turun (2-1) untuk membalik
- **`" " * (3-i)`** → Spasi untuk center alignment
- **`"*" * (2*i-1)`** → Jumlah bintang (1,3,5,3,1)

**Konsep Penting:**
- Inner loop akan **reset dan complete** untuk setiap iterasi outer loop
- Pattern complexity meningkat exponentially dengan nested loops

### ⚡ Break dan Continue
Kontrol alur loop dengan `break` (keluar) dan `continue` (lanjut):

````python
# Contoh break - berhenti saat menemukan angka 7
print("Mencari angka 7:")
angka_list = [1, 3, 5, 7, 9, 11, 13]
for angka in angka_list:
    if angka == 7:
        print(f"Ditemukan! Angka {angka}")
        break
    print(f"Bukan 7: {angka}")

print("\n" + "="*20)

# Contoh continue - lewati angka genap
print("Hanya tampilkan angka ganjil:")
for i in range(1, 11):
    if i % 2 == 0:  # jika genap
        continue    # lewati, lanjut iterasi berikutnya
    print(f"Angka ganjil: {i}")

print("\n" + "="*20)

# Kombinasi - cari bilangan prima sederhana
print("Bilangan prima dari 2-20:")
for num in range(2, 21):
    is_prima = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prima = False
            break
    if is_prima:
        print(num, end=" ")
print()
````

### 🔍 Analisis Kode Break & Continue:

**1. Break Example:**
- **`break`** → Keluar dari loop secara paksa
- Loop **berhenti total** ketika kondisi terpenuhi
- Sisa iterasi **diabaikan**
- Useful untuk **early termination** ketika target ditemukan

**2. Continue Example:**  
- **`continue`** → Skip sisa kode di iterasi saat ini, lanjut ke iterasi berikutnya
- **`i % 2 == 0`** → Modulo operator untuk cek genap (sisa bagi 2 = 0)
- Useful untuk **filtering** atau skip certain conditions

**3. Prime Number Algorithm:**
- **Nested loop** dengan break untuk efficiency
- **`int(num**0.5) + 1`** → Cek hingga akar kuadrat saja (optimization)
- **`is_prima = False`** + `break` → Stop checking ketika ditemukan pembagi
- **Boolean flag** (`is_prima`) untuk track status

**Perbedaan Kunci:**
- **`break`** = Keluar dari loop
- **`continue`** = Skip ke iterasi berikutnya
