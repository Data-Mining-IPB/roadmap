# 3. 🔄 Casting (Konversi Tipe Data)

Casting adalah proses mengubah tipe data dari satu jenis ke jenis lainnya:
- `int()` - mengubah ke integer
- `float()` - mengubah ke float
- `str()` - mengubah ke string
- `bool()` - mengubah ke boolean

````python
# Contoh casting
print("=== Casting ke Integer ===")
angka_float = 3.14
angka_string = "42"
print(f"float {angka_float} ke int: {int(angka_float)}")
print(f"string '{angka_string}' ke int: {int(angka_string)}")

print("\n=== Casting ke Float ===")
angka_int = 10
angka_string = "3.14"
print(f"int {angka_int} ke float: {float(angka_int)}")
print(f"string '{angka_string}' ke float: {float(angka_string)}")

print("\n=== Casting ke String ===")
angka = 123
boolean = True
print(f"int {angka} ke string: '{str(angka)}'")
print(f"bool {boolean} ke string: '{str(boolean)}'")

print("\n=== Casting ke Boolean ===")
print(f"int 0 ke bool: {bool(0)}")
print(f"int 5 ke bool: {bool(5)}")
print(f"string '' ke bool: {bool('')}")
print(f"string 'hello' ke bool: {bool('hello')}")
````

### 📝 Contoh Penerapan Casting

````python
# Contoh kasus: Menghitung total belanja
harga_barang1 = "15000"  # data dari input user (string)
harga_barang2 = "25000"  # data dari input user (string)
harga_barang3 = "12500"  # data dari input user (string)

# Konversi ke integer untuk perhitungan
total = int(harga_barang1) + int(harga_barang2) + int(harga_barang3)
print(f"Total belanja: Rp {total:,}")

# Menghitung diskon
diskon_persen = 15
diskon_rupiah = total * (diskon_persen / 100)
total_bayar = total - diskon_rupiah

print(f"Diskon {diskon_persen}%: Rp {diskon_rupiah:,.0f}")
print(f"Total yang harus dibayar: Rp {total_bayar:,.0f}")
````

# 📖 Print Formatting (Format Output)

Kadang kita ingin menampilkan output yang lebih rapi dan terformat. Python menyediakan beberapa cara mudah:

## 1. **Print biasa** (standar)
## 2. **f-string** (paling mudah dan modern)
## 3. **format()** method (alternatif lain)

````python
# Contoh Perbedaan Print Formatting
nama = "Budi"
umur = 25
tinggi = 175.5

print("=== 1. PRINT BIASA ===")
print("Nama:", nama)
print("Umur:", umur, "tahun")
print("Tinggi:", tinggi, "cm")

print("\n=== 2. F-STRING (PALING MUDAH) ===")
print(f"Nama: {nama}")
print(f"Umur: {umur} tahun")
print(f"Tinggi: {tinggi} cm")

# F-string dengan formatting angka
print(f"Tinggi: {tinggi:.1f} cm")  # 1 angka di belakang koma
print(f"Umur tahun depan: {umur + 1} tahun")  # bisa operasi langsung

print("\n=== 3. FORMAT() METHOD ===")
print("Nama: {}".format(nama))
print("Umur: {} tahun".format(umur))
print("Tinggi: {:.1f} cm".format(tinggi))

print("\n=== PERBANDINGAN OUTPUT ===")
print("Print biasa: Nama:", nama, "- Umur:", umur)
print(f"F-string: Nama: {nama} - Umur: {umur}")
print("Format: Nama: {} - Umur: {}".format(nama, umur))
````

````python
# Contoh Praktis F-String
print("\n=== CONTOH PRAKTIS F-STRING ===")

# Memformat angka
harga = 25000
print(f"Harga: Rp {harga}")
print(f"Harga dengan koma: Rp {harga:,}")  # 25,000

# Memformat desimal
nilai = 85.6789
print(f"Nilai: {nilai}")
print(f"Nilai (1 desimal): {nilai:.1f}")  # 85.7
print(f"Nilai (2 desimal): {nilai:.2f}")  # 85.68

# Contoh untuk tabel sederhana
print(f"\n{'Nama':<10} | {'Umur':<5} | {'Nilai':<6}")
print(f"{'Andi':<10} | {20:<5} | {85.5:<6.1f}")
print(f"{'Budi':<10} | {22:<5} | {92.0:<6.1f}")

# Operasi dalam f-string
tahun_sekarang = 2024
tahun_lahir = tahun_sekarang - umur
print(f"\n{nama} lahir tahun {tahun_lahir}")
print(f"Tahun depan umur {nama} akan {umur + 1} tahun")
````

````python
# Kapan Pakai Yang Mana?
print("\n=== KAPAN PAKAI YANG MANA? ===")

# 1. Print biasa - untuk output sederhana
print("Hasil:", 10 + 5)
print("Status:", "Berhasil")

# 2. F-string - untuk gabungan teks dan variabel (PALING DIREKOMENDASIKAN)
skor = 95
print(f"Skor kamu: {skor}/100")
print(f"Persentase: {skor}%")

# 3. Format() - alternatif lain (kurang praktis)
print("Skor kamu: {}/100".format(skor))

print("\n=== KESIMPULAN ===")
print("✅ Gunakan f-string untuk sebagian besar kasus")
print("✅ Print biasa untuk output yang sangat sederhana")
print("✅ Format() hanya jika ada alasan khusus")
````

### 📝 Contoh Penerapan Print Formatting

Membuat kartu identitas sederhana:

````python
# Aplikasi: Kartu Identitas Sederhana
nama = "Siti Aminah"
kelas = "XII IPA 1"
nomor = 12345
nilai_rata = 87.5

print("="*30)
print("    KARTU PELAJAR")
print("="*30)
print(f"Nama  : {nama}")
print(f"Kelas : {kelas}")
print(f"No    : {nomor}")
print(f"Rata² : {nilai_rata:.1f}")
print("="*30)

# Contoh dengan print biasa (lebih berantakan)
print("\nDengan print biasa:")
print("Nama :", nama)
print("Kelas:", kelas)
print("No   :", nomor)
print("Rata²:", nilai_rata)

# Contoh membuat daftar nilai
print(f"\n{'No':<3} | {'Nama':<12} | {'Nilai':<5}")
print("-" * 25)
print(f"{1:<3} | {'Andi':<12} | {85:<5}")
print(f"{2:<3} | {'Budi':<12} | {92:<5}")
print(f"{3:<3} | {'Cici':<12} | {78:<5}")
````

### 📋 Format Penting untuk Pemula

| Format | Fungsi | Contoh |
|--------|--------|--------|
| `f"{nama}"` | Masukkan variabel | "Budi" |
| `f"{angka:.1f}"` | 1 angka desimal | 85.5 |
| `f"{angka:.2f}"` | 2 angka desimal | 85.50 |
| `f"{angka:,}"` | Pemisah ribuan | 1,000 |
| `f"{teks:<10}"` | Rata kiri, lebar 10 | "hello     " |
| `f"{teks:>10}"` | Rata kanan, lebar 10 | "     hello" |

### 🎯 Tips Print Formatting untuk Pemula
1. **Gunakan f-string** - paling mudah dan modern
2. **Mulai dari yang sederhana** - cukup `f"{variabel}"`
3. **Pelajari bertahap** - tambahkan formatting sedikit demi sedikit
4. **Praktek langsung** - coba variasi yang berbeda

# 5. ✅ Booleans (True/False)

Boolean adalah tipe data yang hanya memiliki dua nilai: `True` atau `False`. Berguna untuk kondisi dan logika program.

````python
# Boolean literals
benar = True
salah = False

print(f"Benar: {benar} (tipe: {type(benar)})")
print(f"Salah: {salah} (tipe: {type(salah)})")

# Boolean dari perbandingan
print("\n=== Perbandingan ===")
angka1 = 10
angka2 = 5
print(f"{angka1} > {angka2}: {angka1 > angka2}")
print(f"{angka1} < {angka2}: {angka1 < angka2}")
print(f"{angka1} == {angka2}: {angka1 == angka2}")
print(f"{angka1} != {angka2}: {angka1 != angka2}")

# Boolean dari fungsi built-in
print("\n=== Fungsi Boolean ===")
teks = "Python"
print(f"'{teks}'.isalpha(): {teks.isalpha()}")
print(f"'{teks}'.isdigit(): {teks.isdigit()}")
print(f"'{teks}'.islower(): {teks.islower()}")
print(f"'{teks}'.isupper(): {teks.isupper()}")
````

### 📝 Contoh Penerapan Booleans

````python
# Sistem validasi data mahasiswa
nama = "Budi Santoso"
umur = 20
ipk = 3.5
sudah_bayar_spp = True
semester = 6

# Validasi data
nama_valid = len(nama) > 0 and nama.replace(' ', '').isalpha()
umur_valid = 17 <= umur <= 30
ipk_valid = 0.0 <= ipk <= 4.0
bisa_ikut_ujian = sudah_bayar_spp and semester >= 1
layak_beasiswa = ipk >= 3.5 and sudah_bayar_spp

print("=== VALIDASI DATA MAHASISWA ===")
print(f"Nama: {nama}")
print(f"Umur: {umur}")
print(f"IPK: {ipk}")
print(f"Status SPP: {'Lunas' if sudah_bayar_spp else 'Belum Lunas'}")
print(f"Semester: {semester}")

print("\n=== HASIL VALIDASI ===")
print(f"Nama valid: {nama_valid}")
print(f"Umur valid: {umur_valid}")
print(f"IPK valid: {ipk_valid}")
print(f"Bisa ikut ujian: {bisa_ikut_ujian}")
print(f"Layak beasiswa: {layak_beasiswa}")

# Status akhir
data_lengkap = nama_valid and umur_valid and ipk_valid
print(f"\nData lengkap dan valid: {data_lengkap}")
````

# 6. ⚡ Operators (Operator)

Python memiliki berbagai jenis operator untuk melakukan operasi pada variabel dan nilai.

````python
# Operator Aritmatika
print("=== OPERATOR ARITMATIKA ===")
a, b = 15, 4
print(f"a = {a}, b = {b}")
print(f"Penjumlahan (+): {a} + {b} = {a + b}")
print(f"Pengurangan (-): {a} - {b} = {a - b}")
print(f"Perkalian (*): {a} * {b} = {a * b}")
print(f"Pembagian (/): {a} / {b} = {a / b}")
print(f"Pembagian bulat (//): {a} // {b} = {a // b}")
print(f"Modulus (%): {a} % {b} = {a % b}")
print(f"Pangkat (**): {a} ** {b} = {a ** b}")

# Operator Perbandingan
print("\n=== OPERATOR PERBANDINGAN ===")
x, y = 10, 20
print(f"x = {x}, y = {y}")
print(f"x == y: {x == y}")
print(f"x != y: {x != y}")
print(f"x < y: {x < y}")
print(f"x > y: {x > y}")
print(f"x <= y: {x <= y}")
print(f"x >= y: {x >= y}")

# Operator Logika
print("\n=== OPERATOR LOGIKA ===")
p, q = True, False
print(f"p = {p}, q = {q}")
print(f"p and q: {p and q}")
print(f"p or q: {p or q}")
print(f"not p: {not p}")
print(f"not q: {not q}")
````

### 📝 Contoh Penerapan Operators

````python
# Operator aritmatika dan logika tanpa if/elif
a = 12
b = 7

# Aritmatika
jumlah = a + b
selisih = a - b
perkalian = a * b
pembagian = a / b
modulus = a % b
pangkat = a ** b

# Perbandingan
lebih_besar = a > b
sama_dengan = a == b

# Logika
keduanya_positif = (a > 0) and (b > 0)
salah_satu_genap = (a % 2 == 0) or (b % 2 == 0)
bukan_genap = not (a % 2 == 0)

print("Jumlah:", jumlah)
print("Selisih:", selisih)
print("Perkalian:", perkalian)
print("Pembagian:", pembagian)
print("Modulus:", modulus)
print("Pangkat:", pangkat)
print("a > b:", lebih_besar)
print("a == b:", sama_dengan)
print("Keduanya positif:", keduanya_positif)
print("Salah satu genap:", salah_satu_genap)
print("Bukan genap (a):", bukan_genap)
````

## 11. 📚 Dictionaries

Dictionary adalah pasangan **kunci–nilai**, diapit oleh `{key: value}`.

````python
# Contoh Dictionary
mahasiswa = {'nama': 'Ani', 'umur': 21, 'jurusan': 'TI'}

print(mahasiswa)
mahasiswa['ipk'] = 3.75
print(mahasiswa['jurusan'])
````

### 📝 Contoh Penerapan Dictionaries

Menyimpan data buku:

````python
# Data buku
data_buku = {'judul': 'Python for Beginners', 'penulis': 'Budi', 'tahun': 2023}
for k, v in data_buku.items():
    print(f"{k}: {v}")
````
